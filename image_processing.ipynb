### Basics of Image Processing

```python
import numpy as np
from skimage import io, img_as_float
from matplotlib import pyplot as plt
```

```python
img = io.imread('images/meme.jpg')

print(img.shape)
print(img.dtype)
print(img.min(), img.max())

plt.imshow(img)
```

```python
img2 = img_as_float(img)

print(img2.shape)
print(img2.dtype)
print(img2.min(), img2.max())

img2[450:710, 410:460, :] = [0, 255, 0]
plt.imshow(img2)
```

```python
import PIL

pil_image = PIL.Image.open('images/meme.jpg')

print(type(pil_image)) 
print(pil_image.format) 
print(pil_image.mode) 
pil_image.show()
```

```python
pil_image1 = np.asarray(pil_image)
print(type(pil_image1))
```

```python
import matplotlib.image as mpimg

mp_image = mpimg.imread('images/meme.jpg')
print(type(mp_image))
print(mp_image)
plt.imshow(mp_image)
```

```python
print(mp_image.dtype)
print(mp_image.min(), mp_image.max())
```

```python
# astype converts pixels to float64
mp_image2 = mp_image.astype(np.float64)
print(mp_image2.dtype)
print(mp_image2.min(), mp_image2.max())
print()

# img_as_float converts pixels to float64 and normalizes them to [0, 1]
mp_image3 = img_as_float(mp_image)
print(mp_image3.dtype)
print(mp_image3.min(), mp_image3.max())
```

```python
import cv2

grey_img = cv2.imread("images/meme.jpg", 0)
color_img = cv2.imread("images/meme.jpg", 1)

# images opened using cv2 are numpy arrays
print(type(grey_img)) 
print(type(color_img)) 

cv2.imshow("pic", grey_img)
cv2.imshow("color pic", color_img)

cv2.waitKey(0)          
cv2.destroyAllWindows() 
```

### Image Preprocessing Operations

```python
import cv2
import numpy as np
```

```python
img = cv2.imread("images/meme.jpg")
```

```python
# Denoising

kernel = np.ones((3, 3), np.float32) / 9

custom_filter = cv2.filter2D(img, -1, kernel)
img_blur = cv2.blur(img, (3, 3))
img_gaussian = cv2.GaussianBlur(img, (3, 3), 0)
img_median = cv2.medianBlur(img, 3)
img_bilateral = cv2.bilateralFilter(img, 9, 75, 75)

cv2.imshow("Original", img)
cv2.imshow("Custom", custom_filter)
cv2.imshow("Blur", img_blur)
cv2.imshow("Gaussian", img_gaussian)
cv2.imshow("Median", img_median)
cv2.imshow("Bilateral", img_bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
